# Convert color image to gray scale

```python
from scipy import misc
import matplotlib
%matplotlib inline
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import numpy as np
import os
from pylab import imread, imshow, gray, mean
gray()



img=plt.imread('C:/Users/Sindhu-Phani/Documents/13_Oct/ML_ANNOTATOR/Annotator/cropped/2011_09_26_drive_0096_sync_0185_c414w125h.png')
print(type(img))
print(img.shape)
#print(img)

def to_grayscale(im, weights = np.c_[0.2989, 0.5870, 0.1140]):
    """
    Transforms a colour image to a greyscale image by
    taking the mean of the RGB values, weighted
    by the matrix weights
    """
    tile = np.tile(weights, reps=(im.shape[0],im.shape[1],1))
    return np.sum(tile * im, axis=2)

imgGray = to_grayscale(img)

imgplot = plt.imshow(imgGray)
hist, bin_edges = np.histogram(imgGray.ravel(),bins=16,range=(0.0,1.0))
print(type(hist))
print(len(hist))
print(hist)
print(type(bin_edges))
print(len(bin_edges))
print(bin_edges)
print("sum is")
print(hist.sum())
```
